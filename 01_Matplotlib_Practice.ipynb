```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
price = [48000,54000,57000,49000,47000,45000]
year = [2015,2016,2017,2018,2019,2020]
```

```python
plt.plot(year,price)
```

```python
# On real world dataset
batsman = pd.read_csv('./Dataset/datasets-session-23/sharma-kohli.csv')
```

```python
batsman.head()
```

```python
# I wanna plot k har season mai inn dono ny kitnay runs banaye hai
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
# Add lebals, like: title, xlabel, ylabel
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.plot(batsman['index'],batsman['V Kohli'])
plt.title('Rohit Vs Kohli Batting Comparison')
plt.xlabel('year')
plt.ylabel('runs')
```

```python
# How to change the colors of the lines
plt.plot(batsman['index'],batsman['RG Sharma'], color = 'Black')
plt.plot(batsman['index'],batsman['V Kohli'], color = 'Green')
plt.title('Rohit Vs Kohli Batting Comparison')
plt.xlabel('year')
plt.ylabel('runs')
```

```python
# How to change the width, and style of the line
plt.plot(batsman['index'],batsman['RG Sharma'], color = 'Black', linestyle= 'solid', linewidth = 3)
plt.plot(batsman['index'],batsman['V Kohli'], color = 'Green', linestyle='dashdot', linewidth = 2)
plt.title('Rohit Vs Kohli Batting Comparison')
plt.xlabel('year')
plt.ylabel('runs')
```

```python
# How to add the marker and how to change the marker size
plt.plot(batsman['index'],batsman['RG Sharma'], color = 'Black', linestyle= 'solid', linewidth = 3, marker = 'o', markersize = 10)
plt.plot(batsman['index'],batsman['V Kohli'], color = 'Green', linestyle='dashdot', linewidth = 2, marker = '+', markersize= 15)
plt.title('Rohit Vs Kohli Batting Comparison')
plt.xlabel('year')
plt.ylabel('runs')
```

```python
# How to add legend. First we have to add the label of each player and then will call the legend()
plt.plot(batsman['index'],batsman['RG Sharma'], color = 'Black', linestyle= 'solid', linewidth = 3, marker = 'o', markersize = 10, label='Virat_Kohli')
plt.plot(batsman['index'],batsman['V Kohli'], color = 'Green', linestyle='dashdot', linewidth = 2, marker = '+', markersize= 15, label='Rohit_Sharma')
plt.title('Rohit Vs Kohli Batting Comparison')
plt.xlabel('year')
plt.ylabel('runs')
plt.legend(loc = 'best') # loc ko ham change b kar sakty hai like: loc = 'upper right', 'upper left' etc
```

```python
# How to give limit to each object
price = [48000,54000,57000,49000,47000,45000,4500000] # There is an  outlier due to which we are not able to figure out the earning from 2015_2020
year = [2015,2016,2017,2018,2019,2020,2021]
plt.plot(year,price)
plt.ylim(0,490000)
```

```python
# Add Grid
plt.plot(batsman['index'],batsman['RG Sharma'], color = 'Black', linestyle= 'solid', linewidth = 3, marker = 'o', markersize = 10, label='Virat_Kohli')
plt.plot(batsman['index'],batsman['V Kohli'], color = 'Green', linestyle='dashdot', linewidth = 2, marker = '+', markersize= 15, label='Rohit_Sharma')
plt.title('Rohit Vs Kohli Batting Comparison')
plt.xlabel('year')
plt.ylabel('runs')
plt.legend(loc = 'best')
plt.grid()
```

```python
# show() # Sometimes plot kabi show nahe hota tu ham plt.show() ka use karty hai
plt.plot(batsman['index'],batsman['RG Sharma'], color = 'Black', linestyle= 'solid', linewidth = 3, marker = 'o', markersize = 10, label='Virat_Kohli')
plt.plot(batsman['index'],batsman['V Kohli'], color = 'Green', linestyle='dashdot', linewidth = 2, marker = '+', markersize= 15, label='Rohit_Sharma')
plt.title('Rohit Vs Kohli Batting Comparison')
plt.xlabel('year')
plt.ylabel('runs')
plt.legend(loc = 'best')
plt.grid()
plt.show()
```

# Bar Chart:-
* Operates on Bivaruate Analyssis
* Can work on Numerical and x = Categorical but most likely (numerical is aggregate())
* Use_Case:- Aggregate Analysis of Groups. {Yani har group ka kitna percentage male hai ya female}

```python
# simple bar chart
children = [10,20,40,10,30]
colors = ['red','blue','green','yellow','pink']

plt.bar(colors, children, color = 'black')
```

```python
# Horizental Bar Chart
plt.barh(colors, children, color='black')
```

```python
df = pd.read_csv('./Dataset/datasets-session-23/batsman_season_record.csv')
```

```python
plt.bar(np.arange(df.shape[0]) - 0.2,df['2015'],width=0.2,color='yellow')
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2,color='red')
plt.bar(np.arange(df.shape[0]) + 0.2,df['2017'],width=0.2,color='blue')

plt.xticks(np.arange(df.shape[0]), df['batsman'])

plt.show()
```

```python
# plt.xticks(rotation='vertical)
children = [10,20,40,10,30]
colors = ['red red red red red red','blue blue blue blue','green green green green green','yellow yellow yellow yellow ','pink pinkpinkpink']

plt.bar(colors,children,color='black')
plt.xticks(rotation='vertical')
plt.show()
```

```python
# Stacked Chart
plt.bar(df['batsman'], df['2017'], label='2017')
plt.bar(df['batsman'], df['2016'], bottom = df['2017'], label= '2016')
plt.bar(df['batsman'], df['2015'], bottom = (df['2016'] + df['2017']), label='2015')
plt.legend()
plt.show()
```

# Pie Chart:-
* operate on both univariate and bivariate
* categorical and numerical both
* Use_Case:- To Find Contribution on a Standard Scale(100%)

```python
data = [10,20,54,65,34,100]
subjects = ['eng','urdu','science','islamyat','chemistry','data_science']
plt.pie(data, labels=subjects)
plt.show()
```

```python
gayle = pd.read_csv('./Dataset/datasets-session-23/gayle-175.csv')
```

```python
plt.pie(gayle['batsman_runs'], labels=gayle['batsman'], autopct='%0.1f%%')
plt.show()
```

```python
# percentage and colors
plt.pie(gayle['batsman_runs'],labels=gayle['batsman'],autopct='%0.1f%%',colors=['blue','green','yellow','pink','cyan','brown'])
plt.show()
```

```python
# explode shadow
plt.pie(gayle['batsman_runs'],labels=gayle['batsman'],autopct='%0.1f%%',explode=[0.3,0,0,0,0,0.1],shadow=True)
plt.show()
```
