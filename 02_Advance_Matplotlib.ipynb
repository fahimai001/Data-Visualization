# Annotation
Using annotion, we can place text on a graph

```python
batter = pd.read_csv('./batter.csv')
```

```python
sample_df = batter.head(100).sample(25, random_state=5)
```

```python
sample_df
```

```python
# Making Scatter plot
plt.figure(figsize=(15,9))
plt.scatter(sample_df['avg'], sample_df['strike_rate'], s=sample_df['runs'])
# Giving Labels(Name) to each Data Points
for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i], sample_df['strike_rate'].values[i], sample_df['batter'].values[i])
```

```python
# I am focusing on those points whos strike_rate is greater than 130 and avg 30. For this I will plot a line among it
# Making Scatter plot
plt.figure(figsize=(15,9))
plt.scatter(sample_df['avg'], sample_df['strike_rate'], s=sample_df['runs'])
# Plotting a line that depicts all those whoes strike_rate is greater than 130 and average is 30
plt.axhline(130, color='red')
plt.axhline(140, color='green')
plt.axvline(30, color='red')
# Giving Labels(Name) to each Data Points
for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i], sample_df['strike_rate'].values[i], sample_df['batter'].values[i])
```

# Subplot

```python
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(15,9), sharex= True)

ax[0].scatter(batter['avg'], batter['strike_rate'], color = 'red')
ax[0].set_title('Avg vs Strike_rate')
ax[0].set_ylabel('Strike_rate')
ax[1].scatter(batter['avg'], batter['runs'], color = 'green')
ax[1].set_xlabel('Avg')
ax[1].set_title('Runs vs Avg')
ax[1].set_ylabel('Runs')
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(15,9))

ax[0,0].scatter(batter['avg'], batter['strike_rate'], color = 'red')
ax[0,0].set_title('Avg vs Strike_Rate')
ax[0,0].set_xlabel('Average')
ax[0,1].scatter(batter['avg'], batter['runs'], color = 'green')
ax[1,0].hist(batter['avg'])
ax[1,1].hist(batter['runs'])
```

# 3D Scatter plots

```python
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
```

```python
xx, yy = np.meshgrid(x,y)
```

```python
z = xx**2 + yy**2
```

```python
fig = plt.figure(figsize=(15,9))

ax = plt.subplot(projection='3d')

p=ax.plot_surface(xx,yy,z, cmap = 'viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.cos(yy)
fig = plt.figure(figsize=(15,9))

ax = plt.subplot(projection='3d')

p=ax.plot_surface(xx,yy,z, cmap = 'viridis')
fig.colorbar(p)
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot()

p = ax.contour(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot()

p = ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

# Heatmap

```python
delivery = pd.read_csv('./IPL_Ball_by_Ball_2008_2022.csv')
```

```python
delivery.head()
```

```python
# Using heatmap I wanna find all those ball where a six is occured
temp_df = delivery[delivery['ballnumber'].isin([1,2,3,4,5,6]) & (delivery['batsman_run']==6)] 
```

```python
grid = temp_df.pivot_table(index='overs', columns='ballnumber', values='batsman_run', aggfunc='count')
```

```python
plt.figure(figsize=(20,10))
# imshow() is used to make a plot from grid like data.
plt.imshow(grid)
plt.yticks(delivery['overs'].unique(), list(range(1,21)))
plt.xticks(np.arange(0,6), list(range(1,7)))
plt.colorbar()
```
